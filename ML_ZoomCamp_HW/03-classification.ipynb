# **Q2**

1- Select Numerical Features: Identify which columns in the DataFrame are numerical.

2- Calculate the Correlation Matrix: Use the .corr() method to compute the correlation matrix.

3- Identify the Most Correlated Features: Look for the pair of features with the highest correlation coefficient.

```python
# Select numerical features from the filtered DataFrame
numerical_features = df_filtered.select_dtypes(include=['int64', 'float64'])
print(numerical_features)
```

```python
# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

```python
correlation_pairs = correlation_matrix.unstack()
correlation_pairs = correlation_pairs[correlation_pairs < 1] 
```

```python
# Identify the maximum correlation pair
max_correlation_pair = correlation_pairs.idxmax()
max_correlation_value = correlation_pairs.max()

print(f"The two features with the biggest correlation are: {max_correlation_pair} with a correlation of {max_correlation_value:.2f}")
```
